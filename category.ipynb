```python
import matplotlib.pyplot as plt
import pandas as pd

# Corrected data for plotting based on the user's latest clarification
content = ['Seattle', 'Michigan', 'Bolivia', 'Topeka', 'Ohio', 'Lake Winnebago']
categories = {
    'Capital': [0, 0, 0, 1, 0, 0],
    'City': [1, 0, 0, 1, 0, 0],
    'Country': [0, 0, 1, 0, 0, 0],
    'Lake': [0, 0, 0, 0, 0, 1],
    'State': [0, 1, 0, 0, 1, 0],
    'Spanish spoken': [0, 0, 1, 0, 0, 0],
    'North America': [1, 1, 0, 1, 1, 1]
}

# Create a DataFrame for easier plotting
df_plot = pd.DataFrame(categories, index=content)

# Plot stacked bar chart
ax = df_plot.plot(kind='bar', stacked=True, figsize=(10, 7), colormap='tab20')
plt.title('Content Categorization')
plt.xlabel('Content')
plt.ylabel('Category Count')
plt.xticks(rotation=45, ha="right")
plt.legend(title="Categories", bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
